```python
def total_costs(item_count, item_cost):
    return item_count * item_cost

print ("hello!")
num1  = input ("How many items did you buy?")
num2  = input ("How much does each item cost?")

# Convert inputs to integers for calculation
num1 = int(num1)
num2 = int(num2)

# Call the function and print its result
print ("The total cost of your purchase is", (total_costs(num1, num2)))
```
